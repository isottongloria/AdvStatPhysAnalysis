```r
#higest and lowest volume lake
a<-top_n(data,1,Volume.km..) %>% arrange(desc(Volume.km..))
('the lake with the max volume is:')
(a[0])


b<-top_n(data,-1,Volume.km..) %>% arrange(desc(Volume.km..))
('the lake with the min volume is:')
(b[0])


#higest and lowest area (km²) lake
c<-top_n(data,1,Area.km2.) %>% arrange(desc(Area.km2.))
('the lake with the max area is:')
(c[0])

d<-top_n(data,-1,Area.km2.) %>% arrange(desc(Area.km2.))
('the lake with the min area is:')
(d[0])
```

#### 2)
Order the frame with respect to the area and determine the two largest area lakes

```r
df=select( data , Volume.km.. , Area.km2.,Length.km.,Max.depth.m.,Mean_depth.m.) #select only the columns of the professor's assignment
df%>%arrange(desc(Area.km2.)) #order the frame wrt area

#higest and lowest area (km²) lake
e<-top_n(data,2,Area.km2.) %>% arrange(desc(Area.km2.))
('the 2 lakes with the max area are:')
(e[0])
```

#### 3 )
by summing up the areas occpupied by the lakes, determine the area of Scotland covered by water

```r
#df_2=select( data , Area.km2.) #select only the columns with the values of areas
f<-df %>% mutate(cum_area = cumsum(Area.km2.))
f
('The total Scotland s area occupied by lakes is:')
(f[13,'cum_area'])
('Km²')
```

# Exercise 2 - Athletics world’s races

#### 1)  
open the results on the Web, reported as a table and create a tibble that contains all the results

```r
#load the libraries to be used in the exercise 
library (rvest)
library (readr)
library(dplyr)
library(lubridate)
library(ggplot2)
```

```r
men100m_html <- read_html( "http://www.alltime-athletics.com/m_100ok.htm " )
men100m_html %>%html_nodes( xpath = " // pre " ) %>% html_text() ->men100m_list
men100m_tbl <- read_fwf( men100m_list,show_col_types = FALSE )  #read fsw() --> read and import fixed width files
as_tibble(men100m_tbl)
```

#### 2)
the resulting tibble has all data represented as ”characters”. Explore the tidyverse packages
and manipulate the tibble to add a name of the columns and to convert the columns to the proper
format

```r
#rename the columns
colnames(men100m_tbl) <- c("rank", "time", "error?", "name", "nationality","Birth","Position","Location","date")
men100m_tbl
```

```r
# Add column names and convert columns to the proper format
#colnames(men100m_tbl) <- c("Rank", "Time", "Name", "Country", "Date", "Place", "Meet", "City", "State")
men100m_tbl$time <- as.numeric(gsub("[[:alpha:]]", "", men100m_tbl$time))
men100m_tbl$date <- as.Date(men100m_tbl$date, "%Y.%m.%d")
```

```r
men100m_tbl
```

```r
# Plot 1: evolution of the fastest time as a function of date of the races
ggplot(men100m_tbl, aes(x = date, y = time)) +
  geom_line() +
  labs(x = "Date of the Race", y = "Fastest Time (seconds)", title = "Evolution of the Fastest Time in 100m Races")
```

```r
# Plot 2: which country had the most fastest runners for each year?
men100m_tbl %>%
  group_by(year = lubridate::year(date), nationality) %>%
  filter(time == min(time)) %>%
  count(nationality, year) %>%
  slice_max(n, with_ties = FALSE, n = 1) %>%
  ggplot(aes(x = year, y = n, fill = nationality)) +
  geom_col() +
  labs(x = "Year", y = "Number of Fastest Runners", title = "Countries with the Most Fastest Runners in 100m Races by Year")
```
